```python
import pandas as pd
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def confidence_interval(data, confidence=0.95):
    n = len(data)
    m, se = np.mean(data), st.sem(data)
    h = se * st.t.ppf((1 + confidence) / 2., n-1)
    return m, h
```

```python
lag = 512

base = pd.read_pickle('./pkl_datasets/mamiraua_dataset_ACF_' + str(lag) + '.gzip')
cotas = pd.read_csv('./boundary_files/Cotas_HxC_bins_' + str(int(lag)) + '.csv')
noise = pd.read_csv('./coloredNoises/coloredNoises_' + str(int(lag)) + '.csv')
```

## Information Theory Quantifier Signatures

```python
time = np.unique(base['time'])

H_mu = np.empty((0))
H_CI = np.empty((0))

C_mu = np.empty((0))
C_CI = np.empty((0))

t = np.empty(0)
for j in time:
        t = np.append(t,str(j[0:2] + ':' + j[2:4]))
        
        H = base.loc[base['time'] == j]['H']
        H_mu = np.append(H_mu,np.mean(H))
        H_CI = np.append(H_CI,confidence_interval(H)[1]) 

        C = base.loc[base['time'] == j]['C']
        C_mu = np.append(C_mu,np.mean(C))
        C_CI = np.append(C_CI,confidence_interval(C)[1])        
```

```python
plt.figure(figsize=(18,8))
plt.rc('font', size=16)
plt.rc('axes', titlesize=16)


plt.subplots_adjust(hspace = 0.6)


plt.subplot(2,1,1)
plt.plot(H_mu)
plt.bar(np.arange(len(H_mu)), H_mu, width=0.8, yerr=H_CI, capsize=3, tick_label = t, color=(0.2, 0.4, 0.6, 0.6))
plt.xlim([-0.5, 47.5])
plt.ylim([np.min(H_mu)-np.max(H_CI),np.max(H_mu)+np.max(H_CI)])
plt.xlabel('Hours of the day')
plt.ylabel('Entropy [H]')
plt.xticks(rotation=90)
plt.title("a)")
plt.rc('font', size=16)
plt.rc('axes', titlesize=16)


plt.subplot(2,1,2)
plt.plot(C_mu)
plt.bar(np.arange(len(C_mu)), C_mu, width=0.8, yerr=C_CI, capsize=3, tick_label = t, color=(0.2, 0.4, 0.6, 0.6))
plt.xlim([-0.5, 47.5])
plt.ylim([np.min(C_mu)-np.max(C_CI),np.max(C_mu)+np.max(C_CI)])
plt.xlabel('Hours of the day')
plt.ylabel('Complexity [C]')
plt.xticks(rotation=90)
plt.title("b)")
plt.rc('font', size=16)
plt.rc('axes', titlesize=16)

# plt.savefig('./figures/Fig6.eps', format="eps", bbox_inches='tight')
plt.show()
```
